## **1. Array and types of array in numpy**

**Array** is a datatype in numpy which is similar to python's list but uses c directly for contiguous memory storation and is incredibly fast compared to standard list in python. Its very effective for large-scale numerical operations. 

Its features are:
- **Homogeneous**: Array in numpy is homogeneous and fixed-type. That means all the data are of same type(e.g. int64, float64, complex128). 
- **Multi-Dimensional**: Array in numpy is also called **ndarray**, and is the fundamental data structure. It can store data in any number of dimensions (1D, 2D, 3D, or more), making it highly flexible for scientific computing and data analysis.
- **Vectorized-Operations**: Array in numpy supports vectorization, where we can perform element-wise operations (like a+b) at near C-speeds which uses optimized, compiled code, eliminating the need for slow python loops.
- **Broadcasting**: Enables operations between arrays of different shapes in a flexible and efficient way.


### **Types of array**

**Nested array**: Arrays that have arrays as their elements

- It could be of any dimension

##### **Tensor**
Tensor is an mathematical object that generalizes scalars, vectors and matrices to higher dimensions. At its core, tensor is an n-dimensional array of numerical values. Tensor originated from physics where it meant objects that encode multi-directional relationship and transforms predictably under changes. 

In AI/ML, tensor is primarily seen as a multi-dimensional array used to store and manipulate data.

## **2. How to create array in Numpy?**

The array object in Numpy is called `ndarray`. 
We can create `ndarray` using the `array`. <br>

*Note*: We can either pass a list or tuple. Since a set uses hashmap and is inherently unordered, Numpy can't guarantee the dtype or ordering.


```python
import numpy as np

arr1 = np.array((1, 2, 3, 4))
arr2 = np.array([1, 2, 3, 4])

print(arr1)
print(arr2)

print(type(arr1))
print(type(arr2))
```

#### Checking the dimension of array using `ndim` attribute of array.

```python
a = np.array(42)
b = np.array([2, 3, 5, 6, 8])
c = np.array([[8, 9, 10], [2, 6, 9]])

d = np.array([ [[2, 5, 6], [5, 6, 7 ]], [[3, 5, 2], [4, 2, 7]]])
```

```python
print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
```

#### While creating array, we can use `ndmin` to have the array in specific dimension.

```python
a = np.array([1, 2, 3, 4, 5], ndmin=4)
```

```python
print(a)

print(a.ndim)
```

```python
import numpy as np

class Linear:
    def __init__(self, input_dim, output_dim):
        # Initialize weights randomly and biases to zero
        # We divide by sqrt(input_dim) to keep starting variance small (Xavier/He initialization)
        self.W = np.random.randn(input_dim, output_dim) / np.sqrt(input_dim)
        self.b = np.zeros(output_dim)

        # We will store X here during the forward pass so we can use it in the backward pass
        self.X_cache = None

        # Gradients will be stored here
        self.dW = None
        self.db = None

    def forward(self, X):
        """
        X shape: (batch_size, input_dim)
        Returns: Z shape (batch_size, output_dim)
        """
        # Save X for the backward pass
        self.X_cache = X

        # TODO: Calculate Z = XW + b
        Z = X @ self.W + self.b

        return Z

    def backward(self, dZ):
        """
        dZ shape: (batch_size, output_dim)
        Returns: dX shape (batch_size, input_dim)
        """
        X = self.X_cache

        # TODO: Calculate dW, db, and dX based on the formulas above
        self.dW = X.T @ dZ
        self.db = np.sum(dZ, axis=0)
        dX = dZ @ self.W.T

        return dX
```

```python
# stable_softmax()

def naive_softmax(X):
  exps = np.exp(X)
  return exps/np.sum(exps, axis=-1, keepdims=True)


```

```python
naive_softmax([1000, 1001, 1002])
```

```python
def stable_softmax(X):
  exps = np.exp(X - np.max(X, axis=-1, keepdims=True))
  return exps/np.sum(exps, axis=-1, keepdims=True)
```

```python
stable_softmax([1000, 1001, 1002])
```
