```python
import numpy as np
```

```python
rng = np.random.default_rng(seed=42)

X = rng.standard_normal((1, 4))
y = 1.0

W1 = rng.standard_normal((4, 7))
b1 = np.zeros((1, 7)) 

W2 = rng.standard_normal((7, 1))
b2 = np.zeros((1, 1))

```

```python
z1 = X @ W1 + b1 # (1, 4) * (4, 7) = (1, 7)
print(f"z1: {z1}")
a1 = np.maximum(z1, 0)
print(f"a1: {a1}")
z2 = a1 @ W2 + b2 # (1, 7) * (7, 1) = (1, 1)
print(f"z2: {z2}")

loss = (z2 - y)**2
print(f"Loss: {loss}")
```

```python
N = X.shape[0]
dL_dz2 = 2/N * (z2 - y)
print(dL_dz2.shape)
dL_dz2
```

```python
dL_dw2 = a1.T @ dL_dz2
print(dL_dw2)
dL_dw2.shape
```

```python
dL_da1 = dL_dz2 * W2.T
print(dL_da1)
dL_da1.shape
```

```python
dL_dz1 = dL_da1 * (z1 > 0)
print(dL_dz1)
dL_dz1.shape
```

```python
dL_dw1 = X.T @ dL_dz1
print(dL_dw1)
dL_dw1.shape
```
