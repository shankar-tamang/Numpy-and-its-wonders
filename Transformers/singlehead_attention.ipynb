```python

def softmax(Z):
    Z = Z - np.max(Z, axis=-1, keepdims=True)
    return np.exp(Z)/ np.sum(np.exp(Z), axis=-1, keepdims=True)

def scaled_dot_product_attention(Q, K, V, mask=None):

    d_k = Q.shape[-1]
    seq_len = Q.shape[-2]
    scores = Q @ K.transpose(0, 2, 1)
    scaled_scores = scores / np.sqrt(d_k)

    if mask is not None:
        scaled_scores += mask
    weights = softmax(scaled_scores)
    output = weights @ V
    return output, weights 

```

```python
rng = np.random.default_rng(42)
N, seq_len, d_k, d_v = 2, 4, 8, 8

Q = rng.standard_normal((N, seq_len, d_k))
K = rng.standard_normal((N, seq_len, d_k))
V = rng.standard_normal((N, seq_len, d_v))

# build causal mask
mask = np.triu(np.ones((seq_len, seq_len)), k=1)
mask[mask == 1] = -np.inf

output, weights = scaled_dot_product_attention(Q, K, V, mask)

print(output.shape)   # should be (2, 4, 8)
print(weights.shape)  # should be (2, 4, 4)
print(weights[0])     # each row should sum to 1.0
print(weights[0].sum(axis=-1))  # should be [1. 1. 1. 1.]
```
