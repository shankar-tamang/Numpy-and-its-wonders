```python
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2/np.pi) * (x + 0.044715 * x**3)))

def feed_forward(X, W1, b1, W2, b2):
    # step 1: first linear layer
    z1 = X @ W1 + b1
    # step 2: activation
    a1 = gelu(z1)
    # step 3: second linear layer
    return a1 @ W2 + b2

# initialize and test with these
N, seq_len, d_model = 2, 4, 8
d_ff = 4 * d_model   # = 32

rng = np.random.default_rng(42)
X  = rng.standard_normal((N, seq_len, d_model))

W1 = rng.standard_normal((d_model, d_ff))
b1 = np.zeros(d_ff)
W2 = rng.standard_normal((d_ff, d_model))
b2 = np.zeros(d_model)
```
