### **Basic operations**

Arithmetic operators on arrays apply elementwise. A new array is created and filled with the result.

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([1, 2, 3, 4])

arr3 = arr1 + arr2
arr3
```

```python
arr3**2
```

#### **Matrix Multiplications**

```python
arr1 = np.array([
    [1, 2],
    [2, 3]
])

arr2 = np.array([
    [2, 1],
    [8, 2]
])

# Product Operator
arr3 = arr1 * arr2 # Multiplation happens elementwise
# print(arr3)
# [[ 2  2]
#  [16  6]]


# Matrix Product

arr4 = arr1 @ arr2
print(arr4)
# [[18  5]
#  [28  8]]

# Another Product

arr5 = arr1.dot(arr2)
print(arr5)
# [[18  5]
#  [28  8]]
```
