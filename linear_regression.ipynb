```python
rg = np.random.default_rng()

rg.random((3,))
```

```python

class LinearRegression():
    def __init__(self, X, y, iteration):
        self.m = rg.random((1,))
        self.b = rg.random((1, ))
        self.X = X
        self.y = y
        self.predicted_y = None
        self.error = None
        self.loss = None
        self.lr = 0.01
        self.iteration = iteration

    def forward(self):
        self.predicted_y = self.X * self.m + self.b
        self.error = self.y - self.predicted_y
        self.loss = np.mean(self.error**2)
        return self.loss
        

    def backward(self):
        grad_x = np.mean(-2 * self.error * self.X )
        grad_b = np.mean(-2 * self.error)
        self.m =  self.m - self.lr * grad_x
        self.b = self.b - self.lr * grad_b
        

    def train(self):

        for x in range(self.iteration):
            Loss = self.forward()
            print(f"Iteration: {x}, Loss: {Loss} ")
            self.backward()

        print(f"Loss: {self.loss}")
        print(f"Final weights: m: {self.m}, b: {self.b}")
       
```

```python
data = [(1, 2), (3, 4), (5, 9), (2, 3)]
data = np.array(data)
```

```python
X = data[:, 0]
y = data[:, 1]
print(X, y)
```

```python
Lr1 = LinearRegression(X, y, 100)
```
