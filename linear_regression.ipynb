### Prerequisites
1. Linear function
2. Gradient calculation
3. Parameters update

```python
import numpy as np
rg = np.random.default_rng()

rg.random((3,))
```

```python

class LinearRegression():
    def __init__(self, X, y, iteration):
        self.m = rg.random((1,))
        self.b = rg.random((1, ))
        self.X = X
        self.y = y
        self.predicted_y = None
        self.error = None
        self.loss = None
        self.lr = 0.01
        self.iteration = iteration

    def forward(self):
        self.predicted_y = self.X * self.m + self.b
        self.error = self.y - self.predicted_y
        self.loss = np.mean(self.error**2)
        return self.loss
        
    def backward(self):
        grad_x = np.mean(-2 * self.error * self.X )
        grad_b = np.mean(-2 * self.error)
        self.m =  self.m - self.lr * grad_x
        self.b = self.b - self.lr * grad_b
    
    def train(self):

        for x in range(self.iteration):
            Loss = self.forward()
            print(f"Iteration: {x}, Loss: {Loss} ")
            self.backward()

        print(f"Loss: {self.loss}")
        print(f"Final weights: m: {self.m}, b: {self.b}")
       
```

### **Vectorized Implementation** 

```python
class LinearRegression():
    def __init__(self, lr=0.01, iterations=1000):
        self.lr = lr
        self.iterations = iterations
        self.W = None
        self.b = None
        self.loss_history = []

    def _param_init(self, n_features):
          
        self.W = np.random.randn(n_features, 1) * 0.01
        self.b = 0.0


    def forward(self, X):

        predicted_y = X @ self.W + self.b
        return predicted_y
    
    def backward(self, X, y, predicted_y):

        n_samples = X.shape[0]

        error = predicted_y - y

        grad_W = (2 / n_samples) * X.T @ error
        grad_b = (2 / n_samples) * np.sum(error)

        self.W -= self.lr * grad_W
        self.b -= self.lr * grad_b



    def fit(self, X, y):

        if y.ndim == 1:
            y = y.reshpae(-1, 1)

        n_samples, n_features = X.shape
        self._param_init(n_features)

        for i in range(self.iterations):
            predicted_y = self.forward(X)

            loss = np.mean((predicted_y - y)**2)
            self.loss_history.append(loss)

            self.backward(X, y, predicted_y)

            if i % (self.iterations // 10) == 0 or i == self.iterations - 1:
                print(f"Iteration: {i:4d} | Loss: {loss:.4f}")

    def predict(self, X):
        return self.forward(X)


```

```python
if __name__ == "__main__":
    # 1. Create some dummy data (100 houses, 3 features each)
    np.random.seed(42)
    X_train = np.random.rand(100, 3)
    
    # 2. Define the "True" rules of the universe that the model needs to learn
    # True weights: [3.0, 1.5, -2.0], True bias: 5.0
    true_W = np.array([[3.0], [1.5], [-2.0]])
    
    # Generate targets (y) based on the true weights, plus a tiny bit of random noise
    y_train = X_train @ true_W + 5.0 + (np.random.randn(100, 1) * 0.1) 
    
    # 3. Initialize our custom model
    print("Starting Training...")
    model = LinearRegression(lr=0.1, iterations=500)
    
    # 4. Train the model
    model.fit(X_train, y_train)
    
    # 5. Check the results (They should be very close to [3.0, 1.5, -2.0] and 5.0)
    print("\n--- Training Complete ---")
    print("True Weights:      [ 3.0,  1.5, -2.0]")
    print("Learned Weights:\n", np.round(model.W.flatten(), 2))
    print(f"True Bias:         5.0")
    print(f"Learned Bias:      {model.b:.2f}")
```
