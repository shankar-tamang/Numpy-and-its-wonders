```python
X = rng.standard_normal([10, 1])
y = X**2
# print(X)
# print(y)

w1 = rng.standard_normal((1, 3))
b1 = np.zeros((1, 3))

w2 = rng.standard_normal((3, 1))
b2 = np.zeros((1, 1))

lr = 0.01

def relu(Z):
    return np.maximum(Z, 0)



for i in range(1000):
    z1 = X @ w1 + b1
    # print(f"z1: {z1}")
    a1 = relu(z1)
    # print(f"a1: {a1}")
    z2 = a1 @ w2 + b2
    # print(f"z2: {z2}")

    loss = np.mean((z2 - y)**2)
    error = z2 - y

    dL_z2 = 2/X.shape[0] * error
    dL_w2 = a1.T @ dL_z2
    dL_b2 = np.sum(dL_z2, keepdims=True, axis=0)
    dL_a1 = dL_z2 @ w2.T
    dL_z1 = dL_a1 * (z1 > 0)
    dL_w1 = X.T @ dL_z1
    dL_b1 = np.sum(dL_z1, keepdims=True, axis=0)

    # print(f"w2: {w2} \nb2: {b2} \nw1: {w1} \nb1: {b1}")

    w1 -= lr * dL_w1
    b1 -= lr * dL_b1
    w2 -= lr * dL_w2
    b2 -= lr * dL_b2

    if i % (1000 // 10 ) == 0 or i == 999:
        print(f"Iterations: {i:4d} | Loss: {loss:.4f}")


```
