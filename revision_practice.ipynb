```python
# Arraay creations
arr1 = np.array([[1, 2], [2, 1]])

print(arr1.shape)
print(arr1.size)
print(arr1.itemsize)
print(arr1.ndim)
print(arr1.strides)
```

```python

arr2 = np.array(arr1, ndmin=4)
print(arr2)
print(arr2.shape)
```
