```python
w1 = 0.2
b1 = 1
w2 = 0.3
b2 = 0.5

X, y = 0.5, 0.1
```

```python

z1 = X * w1 + b1
print(f"z1: {z1}")
```

```python
a1 = z1 if z1 > 0 else 0
print(f"a1: {a1}")
```

```python
z2 = a1 * w2 + b2
print(f"z2: {z2}")
```

```python
loss = (z2 - y)**2
print(f"Loss: {loss}")
```

```python
dL_dz2 = 2 * (z2 - y)
print(dL_dz2)
```

```python
dL_dw2 = dL_dz2 * a1
print(dL_dw2)
```

```python
dL_db2 = dL_dz2
print(dL_db2)
```

```python
dL_da1 = dL_dz2 * w2
print(dL_da1)
```

```python
dL_dz1 = dL_da1 * (z1 > 0)
print(dL_dz1)
```

```python
dL_dw1 = dL_dz1 * w1
print(dL_dw1)
```

```python
dL_db1 = dL_dz1
print(dL_db1)
```
